### What is kernel principal component analysis (KPCA)
KPCA is a non-linear dimensional reduction method, while PCA is a linear method. To better understand KPCA, we will first map the original dataset to a higher dimensional space with a non-linear mapping function $\phi: \mathbb{R}^{d\times n}\rightarrow\mathbb{R}^{h\times n}$ where $h$ is bigger than $d$. More formally, this process is expressed as

$$X\in \mathbb{R}^{d\times n} \rightarrow \phi(X)\in \mathbb{R}^{h\times n},$$

where 

$$h \gg d.$$

Then, we compute the eigenvalues and eigenvectors of $\phi(X)^T\phi(X)$. The data of reduced dimension $Y:\mathbb{R}^{p\times n}$, $p < d$ ($p$ denotes the number of reduced dimensions) can be derived as 

$$Y=\Sigma_p V_p^T,$$

where $\Sigma_p $ is the top-p singular values of $\phi(X)$, and $V_p$ is the corresponding $p$ eigenvectors of $\phi(X)^T\phi(X)$. This process is the same as dual PCA except that dual PCA reduces dimension directly in the original space.  

It is however very difficult to know which non-linear mapping function $\phi$ can map data to higher dimension correctly. Therefore, we need the kernel trick, using which we do not need to know $\phi$ (or $\phi(X)^T\phi(X)$) explicitly but still can do dimensionality reduction. Let's definte $K$ as a kernel function

$$K(X,X)=\phi(X)^T\phi(X).$$ 

Common choices of $K$ are linear kernel, polynomial kernel and Gaussian kernel. The last one is also known as radial basis function (RBF) kernel. The three kernel functions are respectively given as
$$K_{ij} = <X_i, X_j>$$
$$K_{ij} = (1 + <X_i, X_j>)^n$$
$$K_{ij} = \exp{(-\gamma \|X_i -X_j \|^2)}.$$

If we use the first kernel, then KPCA will become dual PCA. The one that is most popular in literature is the third one - RBF kernel, which we will detail below. 

### How to do KPCA 
Before KPCA, let's first produce a moon dataset and visualise it.

#### Import the moon dataset.

As you can see form above plots, if we project the moon dataset to first two principal components (meaning no dimensionlity reduction) we get a same pattern dataset (second figure) but with a different rotation angel. The moon dataset after dimension reduction (third figure) is not separate well. This is due to the fact PCA is a linear algorithm, while the data in the original moon dataset follows a nonlinear distribution. We therefore need to resort to KPCA. 

Note that, we use `plt.scatter` to show data points and their labels (component `c`).
* `plt.scatter`: a plot function which takes each dimension of the dataset as well as their label as inputs.
* `plt.title`: a function to name your figure.
* `plt.figure`: a function to create a figure.

#### Define the RBF kernel. 

```python
def RBF(X, gamma = 0.1):
    
    d,n = X.shape
    K = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            tmp = X[:,i]-X[:,j]
            K[i,j] = np.exp(-gamma*(np.dot(tmp.T,tmp)))
            
    return K
```

Above, we define the RBF (Gaussian) kernel, which is $K_{ij} = \exp{(-\gamma \|X_i -X_j \|^2)}$, through the function `RBF`
* gamma: A hyperparameter $\gamma$ (default: 0.1) in $K$ that needs tunning manually for different applications.
* `np.exp`: The exponential operation
* `np.zeros`: A function that creates a matrix full of zeros

#### Centering the RBF kernel. 

One property of kernels is they should be centered such that $\sum_{ij}K_{ij}=0$. To do this, we create a centering matrix $H \in R^{n\times n}$
$$H = I-\frac{1}{n}\boldsymbol{e}\boldsymbol{e}^{T},$$
where $\boldsymbol{e}=[1,1,\cdots,1]^{T}\in \mathbb{R}^{n\times 1}$. Then, applying
$$ \bar K = HKH$$
will create us a double centered matrix where both rows and columns have zero means, implying that $\sum_{ij}\bar K_{ij}=0$. 

```python
def centering(K):
   
    num, _ = K.shape
    
    e = np.ones((num,1),np.float32)
    I = np.identity(num) 
    H = I - (1.0/num)*np.dot(e, e.T) 
    K_bar = np.dot(np.dot(H,K),H)
    
    return K_bar
```

* `np.identity`: creates a identity matrix
* `np.ones`: creates a matrix full of ones
* `np.dot`: does matrix multiplication

#### KPCA

Then, we can compute the eigenvalues and eigenvectors of $\bar K$. The data with reduced dimension $Y:\mathbb{R}^{p\times n}$ is given as 

$$Y=\Sigma_p V_p^T,$$

where $\Sigma_p $ is a diagonal matrix that consists of the square root of top-p eigenvalues of $\bar K$, and $V_p$ denotes the corresponding $p$ eigenvectors of $\bar K$.

```python
gamma, p = 10, 2
K_bar = centering(RBF(train_data.T, gamma))

eig_val, eig_vec = np.linalg.eig(K_bar)
eig_val_ind = np.argsort(eig_val)
eig_val_ind = eig_val_ind[::-1]
eig_val_ind = eig_val_ind[0:p]

V_p = eig_vec[:,eig_val_ind]
eig_val_p = eig_val[eig_val_ind]
sigma_p = np.diag(eig_val_p**0.5)

Y = np.dot(sigma_p,V_p.T)
```

* `np.linalg.eig`: a function that computes all eigenvectors and eigenvalues
* `np.argsort`: a sort function that returns the index of the ranked array (from min to max)
* `np.diag`: creates a diagonal matrix that take a vector as input

```python
fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(131)
ax1.scatter(train_data[:, 0], train_data[:, 1], c = train_label)
ax1.set_title("original 2D moon dataset")

ax2 = fig.add_subplot(132)
ax2.scatter(Y[0,:], Y[1,:], c = train_label)
ax2.set_title("moon dataset after applying KPCA with p=2")

y_axis = np.zeros(150)
ax3 = fig.add_subplot(133)
ax3.scatter(Y[0,:], y_axis, c = train_label)
ax3.set_title("moon dataset after applying KPCA with p=1")

plt.show()
```

As you can see form the plots above, if we project the moon dataset to the first two principal components using KPCA we get a different pattern dataset (second figure), which is almost linearly separable. If we then project the moon dataset to the first principal component, we end up with a very well separate dataset, as shown in the third figure.

#### Project out-of-sample (test) data

If we have a few test sample points and want to project it to a lower dimension without re-computing eigenvectors, we can use the following equation
$$y=\Sigma_p^{-1} V_p^T \phi(X)^T \phi(x) = \Sigma_p^{-1} V_p^T K(X,x) .$$

Clearly, we need to know how to compute $K(X,x)$, where $X$ denotes samples in training set and x denotes samples in test set. The following code computes $K(X,x)$

```python
def RBF_test(train, test, gamma = 0.1):
    
    _, n1 = train.shape
    _, n2 = test.shape
    K = np.zeros((n1,n2))
    for i in range(n1):
        for j in range(n2):
            tmp = train[:,i] - test[:,j]
            K[i,j] = np.exp(-gamma*(np.dot(tmp.T,tmp)))        
    return K
```

Do not forget that we need to make sure $\sum_{ij}K_{ij}=0$. To do this, we create a centering matrix $H_1 \in R^{n_1 \times n_1}$
$$H_1 = I-\frac{1}{n}\boldsymbol{e}\boldsymbol{e}^{T}$$
and $H_2 \in R^{n_2 \times n_2}$
$$H_2 = I-\frac{1}{n}\boldsymbol{e}\boldsymbol{e}^{T},$$

where $n_1$ and $n_2$ are numbers of samples in training set and test set, respectively. Then, applying
$$ \bar K = H_1KH_2$$
will create us a double centered matrix where both rows and columns have zero means, implying that $\sum_{ij}\bar K_{ij}=0$. 

```python
def centering_test(K):
   
    n1, n2 = K.shape
    
    e = np.ones((n1,1),np.float32)
    I = np.identity(n1) 
    H1 = I - (1.0/n1)*np.dot(e, e.T) 
    
    e = np.ones((n2,1),np.float32)
    I = np.identity(n2) 
    H2 = I - (1.0/n2)*np.dot(e, e.T) 
    
    K_bar_test = np.dot(np.dot(H1,K),H2)
    
    return K_bar_test
```

Now use the following equation to project out-of-sample data
$$y = \Sigma_p^{-1} V_p^T K(X,x) .$$

```python
gamma = 10
K_bar_test = centering_test(RBF_test(train_data.T, test_data.T, gamma))
sigma_p_inv = np.diag(eig_val_p**(-0.5))
Y_test = np.dot(np.dot(sigma_p_inv,V_p.T),K_bar_test)
```

```python
#plot test data before and after dimenional reduction
fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(131)
ax1.scatter(test_data[:,0], test_data[:,1], c = test_label)
ax1.set_title("original 2D moon test set")

ax2 = fig.add_subplot(132)
ax2.scatter(Y_test[0,:], Y_test[1,:], c = test_label)
ax2.set_title("moon test set after applying KPCA with p=2")

y_axis = np.zeros(50)
ax3 = fig.add_subplot(133)
ax3.scatter(Y_test[0,:], y_axis, c = test_label)
ax3.set_title("moon test set after applying KPCA with p=1")

plt.show()
```

Then we show the data with reduced dimension in 2D space (second figure) and in 1D space (third figure). This is due to the fact that PCA is equivalent to MDS if we use the Euclidean distance as our distance matric.
