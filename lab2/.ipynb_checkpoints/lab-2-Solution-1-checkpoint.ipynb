```python
# Here we import all packages we need.

# This is a package for handling and manipulating data.
import pandas as pd

# This is the standard package for plotting.
from matplotlib import pyplot as plt 

# Seaborn is build on matplotlib.
# It is more comfortable to use than pure matplotlib.
import seaborn as sns

# Numpy helps as handling arrays of numbers
import numpy as np

# This computes the normal distribution's PDF
def normpdf(x, mean, sd):
    var = float(sd)**2
    denom = (2*np.pi*var)**.5
    num = np.exp(-((x)-(mean))**2/(2*var))
    return num/denom
```

## Plotting a Normal Distribution

```python
mi = 100
ma = 300
inc = 1

# This produces an array of numbers between mi and ma, incremented by inc
x = np.arange(mi, ma, inc)

# Lets output the numbers
x
```

```python
mean = 200
std= 20

# This will calculate the pdf values of a Gaussian
y_pdf = normpdf(x, sd=std, mean= mean)

# Lets output the numbers
y_pdf

# If we like, we can modify the numbers by, adding subtrac, etc... 
y_mod = y_pdf*0.4
```

```python
# Now lets plot the distribution
sns.lineplot(x=x, y=y_pdf, lw=2, color='red')
```

```python
# We can overlay it on another plot like this
sns.kdeplot(data = penguins, x = 'flipper_length_mm', bw_adjust = 0.1)
sns.lineplot(x=x, y=y_pdf, lw=2, color='red')


# Lets overlay the modified version as well
sns.lineplot(x=x, y=y_mod, lw=2, color='green')

plt.xlim(150,250)
plt.show()
```

Let $X$ be the random variable representing the `sex` of a penguin, and $S$ the one representing the `species`.
Make a pandas data frame named `jointTable` that corresponds to the estimated (relative frequency) joint probability mass function for these two variables.
Each row should correspond to one entry in the table, i.e., one combination of `sex` and `species`.
The data frame should have one column named `prob` that holds the joint probability and one column named `count` that corresponds to the number of penguins in this category.

To display the joint distribution, you can use `.pivot(columns='species', index='sex', values='prob')` to rearrange the table to the conventional form with rows corresponding to one variable and columns corresponding to the other. See details on the `.pivot` function [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html).
You can visualise this pivoted table using [sns.heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html). Set `annot=True` to display the numbers. Use `vmin`, `vmax` to adjust the min max of the colour map.

```python
# Please implement
```

## Solution

```python
penguins['count']=1
penguins['prob']=1
jointTable=penguins.groupby(['sex','species']).count()
jointTable['prob']=jointTable['count']/jointTable['prob'].sum()
jointTable = jointTable[['count','prob']]
jointTable
```

```python
jointTable = jointTable.reset_index()
jointTable.pivot(columns='species', index='sex', values='prob')
```

# Excercise 7
Assume that the distributions $p(m|X=$Female$, S=$Gentoo$ )$ and $p(m|X=$Male$, S=$Gentoo$ )$ are normal distributions.
Estimate their parameters from the data.
Plot the resulting PDFs overlayed on top of the corresponding histogram/KDE plots for the respective distribution.

Use `plt.xlim(3200,6800)` to fix the x-range.

```python
# Please implement
```

## Solution

```python
# Considering only male and female Gentoos
Male = Gentoo.query('sex == "Male"')
Female = Gentoo.query('sex == "Female"')

# Calculate the mean and std for each
meanMale=Male['body_mass_g'].mean()
stdMale=Male['body_mass_g'].std()

meanFemale=Female['body_mass_g'].mean()
stdFemale=Female['body_mass_g'].std()

# This is just to calculate a sensible min/max for the x values
mini=Gentoo['body_mass_g'].min()
maxi=Gentoo['body_mass_g'].max()

# Produce an array of x values
x = np.arange(mini, maxi, 1)

# Calculate y values for Gaussian PDF for male Gentoo body mass 
yMale = normpdf(x, sd=stdMale, mean= meanMale)
# Calculate y values for Gaussian PDF for female Gentoo body mass 
yFemale = normpdf(x, sd=stdFemale, mean= meanFemale)

# Plot everything
sns.histplot(data = Male, x = 'body_mass_g', hue='sex', 
             common_norm= False, bins = 15, stat='density')
sns.lineplot(x=x, y=yMale, lw=2, color='red')
plt.xlim(3200,6800)
plt.show()

sns.histplot(data = Female, x = 'body_mass_g', hue='sex', 
             common_norm= False, bins = 15, stat='density')
sns.lineplot(x=x, y=yFemale, lw=2, color='red')
plt.xlim(3200,6800)
plt.show()
```

# Excercise 8
Using your joint distribution table from Excercise 3 and your two Gaussian distributions from Excercise 7, how can you compute the PDF for the distribution $p(m| S=$Gentoo$)$?

Plot this computed distribution over the corresponding KDE/Histogram plot.
Use `plt.xlim(3200,6800)` to fix the x-range.

```python
# Please implement
```

## Solution
We first compute $p(m,x|S=$Gentoo$)$.
Then compute $p(m|S=$Gentoo$) = p(m,X=$Male$|S=$Gentoo$) + p(m,X=$Female$|S=$Gentoo$)$, using marginalisation.

### How can we compute $p(m,x|S=$Gentoo$)$?
We can compute $p(m,X=$Male$|S=$Gentoo$) = p(m,|X=$Male$, S=$Gentoo$) p(X=$Male$| S=$Gentoo$)$, we use the product rule. This works, because everything is conditioned on $s$ (see slide 10 in lecture).

The same works for $p(m,X=$Female$|S=$Gentoo$) = p(m,|X=$Female$, S=$Gentoo$) p(X=$Male$| S=$Gentoo$)$.

Note that we allready computed $p(m,|X=$Male$, S=$Gentoo$)$ and  $p(m,|X=$Female$, S=$Gentoo$)$ (the Gaussian distributions) in Exercise 7, so we are only missing $p(X=$Male$| S=$Gentoo$)$ and  $p(X=$Female$| S=$Gentoo$)$.

### How do we compute $p(x| S=$Gentoo$)$?
This is analogue to the conditional distributions computed in Exercise 4. We can compute it form the joint probability table as $p(x|s) = p(x,s) / p(s)$, where $p(x,s)$ and $p(s)$ can be taken/computed from the table.


```python
# Computing the conditional dist p(x|S=Gentoo)

# We first compute p(x|s)
# This is analogue to Exercise 4 
pxUs=jointTable.query('species == "Gentoo"')[['sex','prob']]
pxUs['prob']= pxUs['prob'] / ps.set_index('species').loc['Gentoo']['prob']
print(pxUs)

# What is the probability p(X=Male|S=Gentoo)
pMale = pxUs.set_index('sex').loc['Male']['prob']

# What is the probability p(X=Female|S=Gentoo)
pFemale = pxUs.set_index('sex').loc['Female']['prob']
# We could also compute it like this
# pFemale = 1.0 - pFemale

# Now we compute 𝑝(𝑚,𝑥|𝑆=Gentoo) using the product rule
pmMale = yMale * pMale
pmFemale = yFemale * pFemale

# Note that since pMale and pFemale sum to 1,
# ultimately, this ammounts to a weighted average between the two Gaussians
y = pmMale + pmFemale

# Here, we plot everything.
#sns.histplot(data = Gentoo, x = 'body_mass_g', common_norm= False, bins = 25, stat='density')
sns.kdeplot(data = Gentoo, x = 'body_mass_g', common_norm= False, bw_adjust=0.2)


sns.lineplot(x=x, y=y, lw=2, color='red')
plt.xlim(3200,6800)
plt.show()
```
